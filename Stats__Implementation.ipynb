```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# normal distribuition and gaussian distribution
mu, sigma = 0,0.1
s = np.random.normal(mu, sigma, size = 1000)
print("count of dps", len(s))
s
```

```python
plt.hist(s, 30, density = True)
plt.show()
```

```python
# Distribution plot
sns.distplot(s)
```

```python
# Uniform distribuiton : - all the outcomes has equal prob. to occur in range

u = np.random.uniform(1, 0, 1000)
plt.hist(u)
```

```python
# Log normal dist>> whose logarithm is normally distrubuted >> right skew
l = np.random.lognormal(mean = 0, sigma = 0.5, size = 1000)
l = l[np.where(abs(l) <5)]
plt.hist(l, bins = 100)
plt.show()
```

```python
# log normal ---> normal >> log Transformation
sns.distplot(np.log(l))
```

```python
# sqrt transformation
sns.distplot(np.sqrt(l))
```

```python
# T distribution

t = np.random.standard_t(df = 1, size = 10000)
t = t[np.where(abs(t) <10)]
plt.hist(t, bins = 1000)
plt.show()
```

```python
# T dis. with df 10 >> tail is thinner , approx normal dist
t = np.random.standard_t(df = 31, size = 10000)
t = t[np.where(abs(t) <10)]
plt.hist(t, bins = 1000)
plt.show()
```

```python
# F distribution
# dfnum is df for numerator, dfden is df for den
f = np.random.f(dfnum = 1, dfden = 1 ,size = 10000)
f = f[np.where(abs(f) <10)]
plt.hist(f, bins = 1000)
plt.show()
```

```python
f = np.random.f(dfnum = 3, dfden = 10 ,size = 10000)
f = f[np.where(abs(f) <10)]
plt.hist(f, bins = 1000)
plt.show()
```

```python
# 1. Bernouli dist. dist with 1 trial, n bernouli trial is binomial distribuiton

b = np.random.binomial(n= 1, p = 0.5 ,size = 1000)
plt.hist(b)
plt.show()
```

```python
# Binomial dist
b = np.random.binomial(n= 100, p = 0.1 ,size = 1000)
plt.hist(b, bins = 20)
plt.show()
```

```python
# Poisson dist
p = np.random.poisson(lam= 5,size = 1000)
plt.hist(p, bins = 20)
plt.show()
```
